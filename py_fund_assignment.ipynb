```python
# 1 load ARGO data file and (exact file name)
# 2 examine keys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data= np.load("../assignments/argo_float_4901412.npz")
data

list(data.keys())
```

```python
#3, 
Salinity= data["S"]
Temp= data["T"]
Pressure= data["P"]

#4 extract the data into named variable arrays 
level = data["levels"]
lon = data["lon"]
date = data["date"]
lat = data["lat"]




print(S,T,P,level,lon,date,lat)
# 5 print the shapes of a T, S, P and date, lat, lon, level
print(S.shape, T.shape, P.shape, level.shape, lon.shape, date.shape, lat.shape)




```

```python
# 6  plot the first column of T, S, P vs levels
T = data["T"][:, 0]  
S = data["S"][:, 0]  
P = data["P"][:, 0]  
levels = data["levels"]

fig = plt.figure(figsize=(10, 8))
ax1= fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)
ax1.plot(T, levels)
ax1.invert_yaxis()
ax2.plot(S, levels)
ax2.invert_yaxis()
ax3.plot(P, levels)
ax3.invert_yaxis()
```

```python
#7  use pcolormesh to plot the series of T and S profiles as a color panel
# flip the yaxis direction the same way as you did above
# add a colorbar

if T.ndim == 1:
    T = T.reshape(-1, 1)
if S.ndim == 1:
    S = S.reshape(-1, 1)

# Transpose
if T.shape[1] == len(levels):
    print("Transposing T and S...")
    T = T.T
    S = S.T

# Check for mismatched levels orientation
if levels[0] > levels[-1]:
    print("Reversing levels array...")
    levels = levels[::-1]
    T = np.flipud(T)
    S = np.flipud(S)

# Raw visualization of T and S with imshow
plt.imshow(T, aspect='auto', origin='lower', cmap='RdBu_r', extent=[0, T.shape[1], levels.min(), levels.max()])
plt.gca().invert_yaxis()
plt.colorbar(label='Temperature (C)')
plt.xlabel('Profile Index')
plt.ylabel('Levels')
plt.title("Temperature Profile")
plt.show()

plt.imshow(S, aspect='auto', origin='lower', cmap='viridis', extent=[0, S.shape[1], levels.min(), levels.max()])
plt.gca().invert_yaxis()
plt.colorbar(label='Salinity (PSU)')
plt.xlabel('Profile Index')
plt.ylabel('Levels')
plt.title("Salinity Profile")
plt.show()
```

```python
#8 get the mean of T and S at each depth level

mean_T = np.mean(T, axis=1)  
mean_S = np.mean(S, axis=1)  


for x, levels in enumerate(level):
    print(f"Mean T at levels {level}: {mean_T[x]}")
    print(f"Mean S at levels {level}: {mean_S[x]}")
```

```python
#9 plot the mean temp and salinity vs level (depth)

# Ensure levels is 1D array with same length as the number of levels in T and S
levels = np.asarray(level).flatten()

plt.figure(figsize=(10, 6))

plt.plot(mean_T, levels, label='Mean Temperature (C)', color='red')
plt.plot(mean_S, levels, label='Mean Salinity (PSU)', color='blue')

plt.gca().invert_yaxis() 
plt.xlabel('Value')
plt.ylabel('Levels')
plt.title('Mean Temperature and Salinity Profiles')
plt.legend()
plt.grid(True)
plt.show()
```
